# ****Clean the dataset****

Drop the null values from all the columns with missing values


```python
df_tmp=df.copy()
```

```python
df_tmp.dropna(subset=['Title'], inplace=True)
df_tmp.dropna(subset=['Price'], inplace=True)
df_tmp.dropna(subset=['User_id'], inplace=True)
df_tmp.dropna(subset=['profileName'], inplace=True)
df_tmp.dropna(subset=['review/summary'], inplace=True)
df_tmp.dropna(subset=['review/text'], inplace=True)
```

```python
df_tmp.shape
```

```python
df_tmp.isnull().sum()
```

Replace any unnecessary character from the textual values with whitespace

```python
contains_semicolon = df_tmp['Title'].str.contains(';')
print(contains_semicolon.sum())
```

```python
df_tmp['Title'] = df_tmp['Title'].str.replace(';', ' ')
```

```python
contains_semicolon = df_tmp['profileName'].str.contains(';')

print(contains_semicolon.sum())
```

```python
df_tmp['profileName'] = df_tmp['profileName'].str.replace(';', ' ')
```

```python
contains_semicolon = df_tmp['review/summary'].str.contains(';')

print(contains_semicolon.sum())
```

```python
df_tmp['review/summary'] = df_tmp['review/summary'].str.replace(';', ' ')
```

```python
contains_semicolon = df_tmp['review/text'].str.contains(';')

print(contains_semicolon.sum())
```

```python
df_tmp['review/text'] = df_tmp['review/text'].str.replace(';', ' ')
```

```python
df_tmp
```

Make the Ids unique

```python
is_duplicate = df_tmp['Id'].duplicated().any()

if is_duplicate:
    print("The values in the 'id' column are not unique.")
else:
    print("The values in the 'id' column are unique.")
```

```python
df_tmp['Id'] = df_tmp['Id'] + '_' + (df_tmp.groupby('Id').cumcount() + 1).astype(str)
```

```python
is_duplicate = df_tmp['Id'].duplicated().any()

if is_duplicate:
    print("The values in the 'id' column are not unique.")
else:
    print("The values in the 'id' column are unique.")
```

```python
df_tmp
```

Replace the quotation marks with whitespaces in the profileName column

```python
df_tmp['profileName'] = df_tmp['profileName'].replace(['["“”„‟”]', "''"], ' ', regex=True)
```

```python
df_tmp['profileName'] = df_tmp['profileName'].replace(["['‘’‛’']", ''], ' ', regex=True)
```

```python
unique_titles_count = df_tmp['Title'].nunique()
print("Number of unique titles:", unique_titles_count)
```
