# ****Clean the dataset****

```python
df_tmp=df.copy()
```

Drop the unnecessary columns

```python
df_tmp.drop(['previewLink','infoLink','image'],axis=1,inplace=True)
```

Remove the box brackets from the author column values

```python
df_tmp['authors'] = df_tmp['authors'].str.replace('[', '').str.replace(']', '')
```

Remove the unnecessary quotation marks from the author column values

```python
df_tmp['authors'] = df_tmp['authors'].str.replace("'",'')
```

Remove the box brackets from the categories column values

```python
df_tmp['categories'] = df_tmp['categories'].str.replace('[', '').str.replace(']', '')
df_tmp['categories'] = df_tmp['categories'].str.replace("'",'')
```

```python
df_tmp
```

Drop the null values from all the columns

```python
df_tmp.dropna(subset=['Title'], inplace=True)
df_tmp.dropna(subset=['description'], inplace=True)
df_tmp.dropna(subset=['authors'], inplace=True)
df_tmp.dropna(subset=['publisher'], inplace=True)
df_tmp.dropna(subset=['publishedDate'], inplace=True)
df_tmp.dropna(subset=['categories'], inplace=True)
df_tmp.dropna(subset=['ratingsCount'], inplace=True)
```

```python
df_tmp
```

```python
df_tmp.isnull().sum()
```

Replace the semicolon character with whitespace

```python
contains_semicolon = df_tmp['Title'].str.contains(';')
counter = contains_semicolon.sum()
print(counter)
```

```python
df_tmp['Title'] = df_tmp['Title'].str.replace(';', ' ')
```

```python
contains_semicolon = df_tmp['description'].str.contains(';')
counter = contains_semicolon.sum()
print(counter)
```

```python
df_tmp['description'] = df_tmp['description'].str.replace(';', ' ')
```

```python
contains_semicolon = df_tmp['authors'].str.contains(';')
counter = contains_semicolon.sum()
print(counter)
```

```python
df_tmp['authors'] = df_tmp['authors'].str.replace(';', ' ')
```
